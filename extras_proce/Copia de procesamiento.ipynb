```python
import json

# Cargar el contenido del archivo JSON
with open('ordenes_r1.json', 'r') as file:
    data = json.load(file)

# Agregar la columna "region" a todas las entradas
for month in data:
    for entry in data[month]:
        entry["region"] = 1

# Guardar el archivo actualizado
with open('ordenes_r1.json', 'w') as file:
    json.dump(data, file, indent=4)


```
